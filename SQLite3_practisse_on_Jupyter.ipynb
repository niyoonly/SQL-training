then I would import pandas and using it to show my dataframe
##  Create database using SQLite


```python
import sqlite3 #IMPORT SQLITE3
connect = sqlite3.connect('something.db') ##CREATE A CONNECTION WITH DATABASE
cursor = connect.cursor() ##CRE
```

```python
cursor.execute("""drop table if exists something""" ) ## DROP A TABLE IF ITS ALREADY EXISTS
```

```python
#CREATING A TABLE
mytable ="""CREATE TABLE IF NOT EXISTS something(ID INTEGER PRIMARY KEY NOT NULL, FNAME VARCHAR(20), LNAME VARCHAR(20), CITY VARCHAR(20), CCODE CHAR(3))"""

cursor.execute(mytable)

print("table is ready")
```

```python
table = """ create table IF NOT EXISTS something(ID INTEGER PRIMARY KEY NOT NULL, FNAME VARCHAR(20), LNAME VARCHAR(20), CITY VARCHAR(20), CCODE CHAR(2));"""
 
cursor.execute(table)
 
print("Table is Ready")
```

```python
#declare sql statement first
statement= """SELECT * FROM something"""
#put statement into cursor object
cursor.execute(statement)

#show the data
print("This is all the data")
my_data = cursor.fetchall()
for show_This in my_data:
    print(show_This)
```

## if i want to show just a few row basically its the same while using .fetchmany() function

```python
statement = """ SELECT * FROM something"""
cursor.execute(statement)

print ("this is some data")
some_data = cursor.fetchmany(3) #in this case i wanted it 3 rows
for show_some_row in some_data:
    print (show_some_row)
```
